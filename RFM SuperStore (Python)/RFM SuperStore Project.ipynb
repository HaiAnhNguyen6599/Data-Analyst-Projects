```python
# Lay bang Order, group by Order ID 
Order = pd.read_excel('Dataset.xlsx', 'Orders')
Order
```

```python
Returned = pd.read_excel('Dataset.xlsx', 'Return')
Returned
```

```python
# Bỏ các giá trị ở bảng Returned 
df3 = pd.merge(Order, Returned, indicator=True, how='outer').query('_merge=="left_only"').drop('_merge', axis=1)
df3
```

```python
# Check nếu cột Returned là null hết thì nghĩa là các Order ID ở bảng Retunred đã bị loại bỏ 
df3['Returned'].isnull().sum()
```

```python
df3.info()
```

```python
# Tinh chi so R của từng khách hàng 
# R = 2017-12-31 - LastPurchaseDate 

# Lấy ngày cuối cùng mà khách đặt hàng, max(Order Date)
df_recency = df3.groupby(by='Customer ID',
                        as_index=False)['Order Date'].max()

# Đặt tên cột lần lượt là Customer ID và LastPurchase Date 
df_recency.columns = ['Customer ID', 'LastPurchaseDate'] 

# tao recent_date = max LastPurchaseDate + 1 Days  
recent_date = df_recency['LastPurchaseDate'].max() + pd.Timedelta(days=1)
#recent_date

# Tao va tinh chi so cot recency 
# x là giá trị của cột LâstPurrhaseDate 
df_recency['Recency'] = df_recency['LastPurchaseDate'].apply(
    lambda x:  (recent_date - x).days)

df_recency
```

```python
# Tính chỉ số F theo từng khách hàng
# F = Count(OrderID)


frequency_df = df3.groupby('Customer ID')['Order ID'].nunique().reset_index()                                    
frequency_df.columns = ['Customer ID', 'Frequency']

frequency_df
```

```python
# Tính chỉ số M của từng khách hàng 
# M = Total Sales of each customer 

monetary_df = df3.groupby(by='Customer ID', as_index=False)['Sales'].sum()
monetary_df.columns = ['Customer ID', 'Monetary']
monetary_df
```

```python
# Tạo bảng RFM bằng cách merge 3 kết quả R F M từ các DataFrame kia 

# Join R và F, key = Customer ID 
df_rf = df_recency.merge(frequency_df, on='Customer ID')

# Join RF với M, key = Customer ID, bỏ cột LastPurchaseDate 
df_rfm = df_rf.merge(monetary_df, on='Customer ID').drop(
    columns='LastPurchaseDate')

df_rfm
```

```python
# Chấm điểm RFM theo Quintile 
df_rfm['R_score'] = pd.qcut(df_rfm['Recency'], q=5, labels=[5,4,3,2,1]) # R càng to điểm càng bé 
df_rfm['F_score'] = pd.qcut(df_rfm['Frequency'], q=5, labels=[1,2,3,4,5])
df_rfm['M_score'] = pd.qcut(df_rfm['Monetary'], q=5, labels=[1, 2, 3, 4, 5])# Dùng rank để 
df_rfm
```

```python
# Tạo RFM score bằng cách Combinevalue các giá trị R F M lại 
df_rfm['RFM Score'] = df_rfm['R_score'].astype(str) +\
                df_rfm['F_score'].astype(str) +\
                df_rfm['M_score'].astype(str)
df_rfm
```

```python
# Import Segment table
df_segment = pd.read_excel('Dataset.xlsx', 'Segmentation')
df_segment
```

```python
# Biến các giá trị trong cột RFM score đang từ dạng string thành list
# Sử dụng .str.splt(',')
df_segment['RFM Score'] = df_segment['RFM Score'].str.split(',')
df_segment
```

```python
# Sử dụng df.explode('column_name') để Unnnest các giá trị RFM score
df_segment = df_segment.explode('RFM Score')
df_segment.sort_values(by='RFM Score')
df_segment 
```

```python
# Convert df_segment['RFM Score'] and rfm_df['RFM Score'] to int type 
df_segment['RFM Score'] = df_segment['RFM Score'].astype('int')
df_rfm['RFM Score'] = df_rfm['RFM Score'].astype('int')

# Merge df_segment and rfm_df
joined = pd.merge(df_rfm, df_segment, on = 'RFM Score', how = 'left')
joined
```

```python
# Double Check null
joined.isnull().sum()
```

### Group Result by Segmentation 

```python
group = joined.groupby('Segment').agg({'Customer ID': 'count',
                                     'Monetary': 'sum',
                                     'Recency': 'mean'}).reset_index()
group.columns = ['Segment','Cust_count','Total_Monetary','Avg_Recency']
group.sort_values(by = 'Cust_count', ascending = False, inplace = True)
group
```

```python
group['count_share'] = group['Cust_count'] / group['Cust_count'].sum()
group['monetary_share'] = group['Total_Monetary'] / group['Total_Monetary'].sum()
group
```

```python
plt.figure(figsize = (20,8))
sns.barplot(data = group, x = 'Segment', y = 'Cust_count')
plt.ylabel('Total Customer')
plt.show()
```

```python
#  Total Customer by Segment
 
fig, ax = plt.subplots(1, figsize = (16,8))

squarify.plot(sizes = group['Cust_count'],
              label = group['Segment'],
              value = [f'{x*100:.2f}%' for x in group['count_share']],
              alpha = .7,
              color = sns.light_palette('green', group.shape[0]),
              bar_kwargs = dict(linewidth = 1.5, edgecolor = "white")
             )
plt.axis('off')
plt.title('RFM Segment of Total Customer', fontsize = 20)
plt.show()
```

```python
# RFM Segment of Total Sales
fig, ax = plt.subplots(1, figsize = (16,8))

squarify.plot(sizes = group['Total_Monetary'],
              label = group['Segment'],
              value = [f'{x*100:.2f}%' for x in group['monetary_share']],
              alpha = .7,
              color = sns.light_palette('green', group.shape[0]),
              bar_kwargs = dict(linewidth = 1.5, edgecolor = "white")
             )
plt.axis('off')
plt.title('RFM Segment of Total Sales', fontsize = 20)
plt.show()
```
